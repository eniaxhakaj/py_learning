# Latin Hypercube Sampling 

* you've got one sample in each row and one in each column 

https://en.wikipedia.org/wiki/Latin_hypercube_sampling

Following example from: https://smt.readthedocs.io/en/latest/_src_docs/sampling_methods/lhs.html. I couldn't use pyDOE because it didn't have multiple dimensions 



```python
import numpy as np 
import matplotlib.pyplot as plt 
from smt.sampling_methods import LHS
```

```python
%matplotlib notebook

plt.rcParams['figure.figsize'] = (6,5)
plt.rcParams['legend.frameon'] = False
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['legend.borderpad'] = 0.1
plt.rcParams['legend.labelspacing'] = 0.1 
plt.rcParams['legend.handletextpad'] = 0.1
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 10
plt.rcParams.update({'figure.autolayout': True})
```

```python
# set the limits for each of the dimensions 
xlimits = np.array([
    [0., 4.],
    [0., 3.],
    [0., 1.],
])

# run the sampling functions
sampling = LHS(xlimits=xlimits)

# get the number of total points sampled 
num = 50
x = sampling(num)

print(x.shape)
```

```python
plt.figure()
plt.title("Tada!")
plt.scatter(x[:, 0], x[:, 1], c = x[:, 2], marker = 'o')
plt.xlabel('x')
plt.ylabel('y')
cbar = plt.colorbar()
cbar.set_label("z")
plt.show()
```
